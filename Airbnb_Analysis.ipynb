1. Neighbourhood Group Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='neighbourhood_group', palette='Set2')
plt.title('Number of Listings per Borough')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

2. Room Type Distribution

```python
sns.countplot(data=df, x='room_type', palette='Set3')
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

3. Price Distribution (up to $500)

```python
sns.histplot(df['price'], bins=100, kde=True, color='teal')
plt.xlim(0, 500)
plt.title('Price Distribution (Capped at $500)')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

 4. Geospatial Map of Listings by Price

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df[df['price']<500], x='longitude', y='latitude', hue='neighbourhood_group', alpha=0.5)
plt.title('NYC Listings on Map (Price < $500)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```

5. Boxplot of Price by Borough

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='neighbourhood_group', y='price')
plt.ylim(0, 500)
plt.title('Price Distribution by Borough (Capped at $500)')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price ($)')
plt.show()
```

6. Availability Histogram

```python
sns.histplot(df['availability_365'], bins=30, kde=False, color='orange')
plt.title('Availability of Listings Throughout the Year')
plt.xlabel('Available Days')
plt.ylabel('Count')
plt.show()
```

7. Top 10 Neighbourhoods by Listing Count

```python
top_neigh = df['neighbourhood'].value_counts().head(10)
top_neigh.plot(kind='barh', color='skyblue')
plt.title('Top 10 Neighbourhoods by Listings')
plt.xlabel('Number of Listings')
plt.gca().invert_yaxis()
plt.show()
```

8. Number of Reviews vs Price

```python
sns.scatterplot(data=df[df['price'] < 500], x='number_of_reviews', y='price', alpha=0.4)
plt.title('Number of Reviews vs Price (Listings < $500)')
plt.xlabel('Number of Reviews')
plt.ylabel('Price ($)')
plt.show()
```

9. Heatmap of Correlations

```python
import numpy as np

numeric_cols = df.select_dtypes(include=np.number)
plt.figure(figsize=(10,8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

10. Average Price per Borough

```python
avg_price = df.groupby('neighbourhood_group')['price'].mean().sort_values()
avg_price.plot(kind='bar', color='mediumseagreen')
plt.title('Average Price by Borough')
plt.ylabel('Average Price ($)')
plt.xlabel('Neighbourhood Group')
plt.xticks(rotation=45)
plt.show()
```
